```python
import math
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams["savefig.dpi"] = 100
import numpy as np
import os
from os import path
import pylab
%matplotlib inline
pylab.rcParams['figure.figsize'] = (20.0, 8.0)
```

```python
def convertAngle(value):
    angle = abs(180/math.pi*math.acos(value))
    if angle > 90:
        angle = 180 - angle
    return angle

def emptyArrayOfLists(size):
    arrayOfLists = [0] * size;
    for n in range(0, size):
        arrayOfLists[n]=[]
    return arrayOfLists
```

We store each files contents in an array of arrays where the data is split by whitespaces. We only read the file if it is not empty using getsize(). We check the file has no 'nan's as they cause problems later.

```python
data = [];
nofiles = 0;
lines = [];
sizes = [0] * 4;
for f in os.listdir('./angmom/'): 
    if path.getsize('./angmom/' + f) > 0:
        if 'e09' in f:
            sizes[0] = sizes[0] + 1;
        elif 'e10' in f:
            sizes[1] = sizes[1] + 1;
        elif 'e11' in f:
            sizes[2] = sizes[2] + 1;
        elif 'e12' in f:
            sizes[3] = sizes[3] + 1;
        nofiles =  nofiles + 1
        nolines = 0
        for line in open('./angmom/' + f, 'r').read().split('\n'):
            linedata = line.split() 
            nolines = nolines + 1
            if len(linedata) == 21 and not 'nan' in line:
                toappend = line.split();
                toappend.append(str(nofiles))
                data.append(toappend)
        lines.append(nolines)
print str(len(data)) + ' galaxies found.'
print sizes;
```

```python
print 'files: ' + str(nofiles) + ' len lines: ' + str(len(lines))
nbins = nofiles
plt.scatter(range(0, nofiles), lines, color='green', alpha=0.5)
plt.show()
```

We take the 10th, 11th and 12th items of data in each of the files data arrays and store them seperately in there own arrays. The data is stored as length ratios so we take the cosine of the ratios to get the angle. If the angle is greater than 90 we take 180 - angle using the convertAngle function defined earlier.

```python
Ldota, Ldotb, Ldotc = [], [], []
Ldotn = [Ldota, Ldotb, Ldotc]
for d in data:  
    for n in range(0, 3):   
        angle = convertAngle(float(d[9+n]))
        Ldotn[n].append(angle)
```
